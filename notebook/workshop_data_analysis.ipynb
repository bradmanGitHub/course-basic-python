### จัดการ Missing data ด้วยการลบออกไป

```python
new_data = data[data.a.notnull()]
```

### ตรวจสอบว่ามาจาก Windows หรือไม่ ?

```python
new_data['os'] = np.where(new_data['a'].str.contains('Windows'),
                          'Windows', 'Not Windows')
```

```python
new_data['os'][:20]
```

## ต้องการดูข้อมูลในแต่ละ Timezone ว่าใช้และไม่ใช้ Windows เท่าไร

```python
by_tz_os = new_data.groupby(['tz', 'os'])
agg_counts = by_tz_os.size().unstack().fillna(0)
agg_counts[:10]
```

### ทำการเรียงข้อมูลจากมากไปน้อย

```python
indexer = agg_counts.sum(1).argsort()
indexer[:15]
```

```python
# Large value
count_subset = agg_counts.take(indexer[-10:])
count_subset
```

```python
type(count_subset)
```

```python
agg_counts.sum(1).nlargest(10)
```

## ทำการ plot graph ด้วย Seaborn

```python
count_subset = count_subset.stack()
```

```python
count_subset.name = 'total'
```

```python
count_subset = count_subset.reset_index()
```

```python
count_subset[:10]
```

```python
import seaborn as sns
%matplotlib notebook
```

```python
count_subset[:10]
```

```python
sns.barplot(x='total', y='tz', hue='os',  data=count_subset)
```

### ข้อมูลในกลุ่มเล็ก ๆ ดูยากมาก ๆ ดังนั้นต้อง nomalize ข้อมูล

```python
def norm_total(group):
    group['normed_total'] = group.total / group.total.sum()
    return group

results = count_subset.groupby('tz').apply(norm_total)
```

```python
results
```

```python
sns.barplot(x='normed_total', y='tz', hue='os',  data=results)
```

### ในการ normalize ข้อมูล แนะนำให้ใช้งาน function transform() กับ groupby()

```python
def norm_total(group):
    g = count_subset.groupby('tz')
    group['normed_total'] = count_subset.total / g.total.transform('sum')
    return group

results2 = count_subset.groupby('tz').apply(norm_total)
```

```python
results2
```

```python
sns.barplot(x='normed_total', y='tz', hue='os',  data=results2)
```
